## Opening and namefiles for output file

```python
import os


pxylem = 70

test_path = "test_data/6_4_2017/"

experiment = test_path.split('/')[-2]


for file_name in os.listdir(test_path): 
    if file_name.endswith('xls'):
        conc, image = file_name.split("_")
        if conc.startswith('0') and len(conc) > 1: 
            conc = float(conc.replace('0','.', 1))
        else: conc = float(conc)
        # read file into pandas
        # get % xylem
        
        print "{0}\t{1}\t{2}".format(pxylem,conc,experiment)
     
```

## Ploting data using Seaborn!

```python
%matplotlib inline 
# rbind experiments

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="ticks")

#load data
xylem_data = pd.read_table("test_data/experment_data.txt")


# Initialize the figure
f, ax = plt.subplots(figsize=(7, 6))
#ax.set_xscale("log")
 
# Plot the orbital period with horizontal boxes
sns.boxplot(x="conc", y="percent_xylem", data=xylem_data,
            whis=np.inf, palette="vlag")

# Add in points to show each observation
sns.swarmplot(x="conc", y="percent_xylem", data=xylem_data,
              size=6, color=".3", hue="experiment", linewidth=0)


# Make the quantitative axis logarithmic
ax.xaxis.grid(True)
ax.set(ylabel="")
sns.despine(trim=True, left=True)
```
